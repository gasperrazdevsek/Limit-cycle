#System:

*   $\dot{x_1} = x_2 - x_1(\frac{1}{3}x_1^2 - 1) $
*   $\dot{x_2}=-\eta x_1$

```
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```
font = {'weight' : 'normal',
        'size'   : 20}

matplotlib.rc('font', **font)
```

# Phase digram

```
eta = 1

def f(X, t):
    x1, x2 = X
    return [x2 - x1*(1/3*x1*x1 - 1), -eta*x1]


N = 20
arr = np.linspace(-4.0, 4.0, N)

xi, yi = np.meshgrid(arr, arr)

t0 = 0

u, v = np.zeros(xi.shape), np.zeros(yi.shape)

for i in range(N):
    for j in range(N):
        x = xi[i, j]
        y = yi[i, j]
        xdot = f([x, y], t0)
        u[i,j] = xdot[0]
        v[i,j] = xdot[1]
     
plt.figure(figsize=(10,8))
plt.quiver(xi, yi, u, v, color='k')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.xlim([-4, 4])
plt.ylim([-4, 4])
```

## Trajectories from chosen starting point



```
eta = 1

def f(t, X):
    x1, x2 = X
    return [x2 - x1*(1/3*x1*x1 - 1), -eta*x1]


a, b = 0, 100
t = np.linspace(a, b, 1000)

plt.figure(figsize=(10,8))
for x0 in [-3, -1, 1, 3]:
    for y0 in [-2, 0, 2]:
        sol = solve_ivp(f, [a, b], [x0, y0], t_eval=t)
        plt.plot(sol.y[0], sol.y[1], ":", color="r")

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.xlim([-4, 4])
plt.ylim([-4, 4])
```

#**Limit cycle for different parameters $\eta$**

```

def plot_limit_cycle(eta, pnt_number):
    def f(t, X):
        x1, x2 = X
        return [x2 - x1*(1/3*x1*x1 - 1), -eta*x1]


    N = 20
    arr = np.linspace(-4.0, 4.0, N)

    xi, yi = np.meshgrid(arr, arr)

    t0 = 0

    u, v = np.zeros(xi.shape), np.zeros(yi.shape)

    for i in range(N):
        for j in range(N):
            x = xi[i, j]
            y = yi[i, j]
            xdot = f(t0,[x, y])
            u[i,j] = xdot[0]
            v[i,j] = xdot[1]
     
    plt.figure(figsize=(10,8))
    plt.quiver(xi, yi, u, v, color='k')

    a, b = 0, pnt_number
    t = np.linspace(a, b, 1000)

    for x0 in [-3, -1, 1, 3]:
        for y0 in [-2, 0, 2]:
            sol = solve_ivp(f, [a, b], [x0, y0], t_eval=t)
            plt.plot(sol.y[0], sol.y[1], ":", color="r")

    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.xlim([-4, 4])
    plt.ylim([-4, 4])
```

```
eta = 0.01
pnt_number = 1000
plot_limit_cycle(eta, pnt_number)
```

```
eta = 0.1
pnt_number = 100
plot_limit_cycle(eta, pnt_number)
```

```
eta = 0.3
pnt_number = 100
plot_limit_cycle(eta, pnt_number)
```
